## Process Data

```sql
CREATE or Alter VIEW v_contacts_updated_zipcode as 

SELECT
    id,
    case
        when len(property_zip) = 4 then concat('0',property_zip)
        else property_zip
    end as zip_code
FROM bronze_lakehouse.dbo.hubspot__contact
where _fivetran_deleted = 0
    and len(property_zip) = 4
```
