Creates view that is used by Census. This view creates an array for each association label so that there is one record per contact.

```sql
CREATE OR ALTER VIEW v_hubspot_association_census_ready AS

SELECT
    c.property_email AS contact_email,
    '[' + STRING_AGG(CASE WHEN a.hubspot_label = 'general_manager' AND a.to_id IS NOT NULL THEN '"' + CAST(a.to_id AS VARCHAR) + '"' END, ',') + ']' AS general_manager_ids,
    '[' + STRING_AGG(CASE WHEN a.hubspot_label = 'owner/principal' AND a.to_id IS NOT NULL THEN '"' + CAST(a.to_id AS VARCHAR) + '"' END, ',') + ']' AS owner_principal_ids,
    '[' + STRING_AGG(CASE WHEN a.hubspot_label = 'never_targeted' AND a.to_id IS NOT NULL THEN '"' + CAST(a.to_id AS VARCHAR) + '"' END, ',') + ']' AS never_targeted_ids,
    '[' + STRING_AGG(CASE WHEN a.hubspot_label = 'sales_manager' AND a.to_id IS NOT NULL THEN '"' + CAST(a.to_id AS VARCHAR) + '"' END, ',') + ']' AS sales_manager_ids,
    '[' + STRING_AGG(CASE WHEN a.hubspot_label = 'finance_manager' AND a.to_id IS NOT NULL THEN '"' + CAST(a.to_id AS VARCHAR) + '"' END, ',') + ']' AS finance_manager_ids,
    '[' + STRING_AGG(CASE WHEN a.hubspot_label = 'unknown' AND a.to_id IS NOT NULL THEN '"' + CAST(a.to_id AS VARCHAR) + '"' END, ',') + ']' AS unknown_ids,
    '[' + STRING_AGG(CASE WHEN a.hubspot_label = 'no_longer_employed' AND a.to_id IS NOT NULL THEN '"' + CAST(a.to_id AS VARCHAR) + '"' END, ',') + ']' AS no_longer_employed_ids
-- FROM [smartsync_lakehouse].[dbo].[hubspot_associations] a
-- INNER JOIN [silver_lakehouse].[dbo].[hubspot__contact] c
--     ON a.from_id = c.id
-- GROUP BY c.property_email
FROM [silver_lakehouse].[dbo].[hubspot__contact] c
LEFT JOIN [smartsync_lakehouse].[dbo].[hubspot_associations] a
    ON a.from_id = c.id
WHERE c.property_email is not null
    AND c.property_email != ''
GROUP BY c.property_email
```
